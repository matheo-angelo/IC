## Downloading the ROI labels for the brain atlas

We are choosing the Harvard-Oxford atlas (with background + 110 brain regions), instead of the default atlas (Craddock 200) for better interpretability.

```python
import pandas as pd
!wget https://fcp-indi.s3.amazonaws.com/data/Projects/ABIDE_Initiative/Resources/ho_labels.csv

ho_labels = pd.read_csv('/content/ho_labels.csv')
ho_labels = ho_labels.rename(columns={'Unnamed: 1' : 'ROI'})
print("Harvard-Oxford atlas labels for ABIDE I:\n")
ho_labels
```

## Function to print biomarker ROI labels

```python
import pickle as pkl

def print_brain_regions(folder_path='/content/BrainGNN/model/', model_name='saved_model'):
  with open(folder_path + model_name + '_explainability_report.pkl', 'rb') as file:
      explanations = pkl.load(file)
  biomarker = ho_labels.iloc[explanations['biomarker']]
  print("Salient brain regions:\n")
  for ROI in biomarker['ROI']:
    print(ROI)
```

## Training BrainGNN

We will train different versions with different hyperparameters, to see the effect in predictive accuracy and interpretability. The hyperparameters to be tested are the following:


*   **Pooling ratio**: the ratio of nodes mantained in the graph in the coarsening (pooling) layer. Default: 0.5.
*   $\lambda_3$: TopK pooling loss in the first message passing layer. Default: 0.1.
*   $\lambda_4$: TopK pooling loss in the second message passing layer. Default: 0.1.
*   $\lambda_5$: Group consistency loss. Default: 0.1.





### Default

```python
%run /content/BrainGNN/03-main.py --save_path="/content/BrainGNN/model" --model_file_name='default' --atlas='ho' --explain=True
print_brain_regions(model_name='default')
```

### Smaller pooling ratio (0.3)

```python
%run /content/BrainGNN/03-main.py --ratio=0.3 --save_path="/content/BrainGNN/model" --model_file_name='ratio' --atlas='ho' --explain=True
print_brain_regions(model_name='ratio')
```

### Bigger TopK Loss regularization (0.5 for both layers)

```python
%run /content/BrainGNN/03-main.py --lamb3=0.5 --lamb4=0.5 --save_path="/content/BrainGNN/model" --model_file_name='topk' --atlas='ho' --explain=True
print_brain_regions(model_name='topk')
```

### Bigger group consistency regularization (0.5)

```python
%run /content/BrainGNN/03-main.py --lamb5=0.5 --save_path="/content/BrainGNN/model" --model_file_name='group_consistency' --atlas='ho' --explain=True
print_brain_regions(model_name='group_consistency')
```

### Smaller pooling ratio (0.3) and bigger TopK Regularization (0.5 for both layers)

```python
%run /content/BrainGNN/03-main.py --ratio=0.3 --lamb3=0.5 --lamb4=0.5 --save_path="/content/BrainGNN/model" --model_file_name='topk_ratio' --atlas='ho' --explain=True
print_brain_regions(model_name='topk_ratio')
```

### Bigger TopK Regularization (0.5 for both layers) and bigger group consistency regularization (0.5)

```python
%run /content/BrainGNN/03-main.py --lamb3=0.5 --lamb4=0.5 --lamb5=0.5 --save_path="/content/BrainGNN/model" --model_file_name='topk_group' --atlas='ho' --explain=True
print_brain_regions(model_name='topk_group')
```
